Имеется два сплава золота с серебром. В первом металлы находятся в отношении 2 : 3. 
Во втором - в отношении 3 : 7. Сколько необходимо взять каждого сплава, 
чтобы получить 8 кг. нового сплава, с отношением золота и серебра 5 : 11?

```python
from sympy import Eq
from sympy import solve
from sympy import symbols

x,y = symbols("x y")

sys_eq = [Eq(x + y, 8), 
          Eq((2 * x / 5 + 3 * y / 10) / (3 * x / 5 + 7 * y / 10), 5/11)]
sol = solve(sys_eq, [x, y])
print(sol)
```

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

eq = Eq((2*x/5+3*(8-x)/10)/(3*x/5+7*(8-x)/10), 5/11)
sol = solve(eq, x)
print([sol[0], 8 - sol[0]])
```
